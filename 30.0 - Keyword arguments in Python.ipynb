```python
## Introduction to Seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = 10,8
%matplotlib inline
```

```python
# Distribuition:  

visl = sns.distplot(stats['Internet users'])
```

```python
visl = sns.distplot(stats['Internet users'], bins = 30)
```

```python
# Boxplot?

vis2 = sns.boxplot(data = stats, x = 'Income Group', y = 'Birth rate')
```

```python
# Seaborn Gallery
```

```python
vis3 = sns.lmplot(data = stats, x = 'Internet users' , y = 'Birth rate', \
                 hue = 'Income Group') # hue means colour
```

```python
vis4 = sns.lmplot(data = stats, x = 'Internet users' , y = 'Birth rate', \
                 hue = 'Income Group', fit_reg = False, height = 8)


# Taking out the lines for the graph be more visible using the 'fit_reg = False'
# the size was the paramter until Python 3.7 but now to change the size you should use 'height'
```

```python
# Keyword Arguments in Python (advanced tutorial)
```

```python
#marker size

vis5 = sns.lmplot(data = stats, x = 'Internet users' , y = 'Birth rate', \
                 hue = 'Income Group', fit_reg = False, height = 8, \
                 scatter_kws = {'s':100})


# scatter_kws allows that you use the plt of matplotlib as param, in the example use s to change the size of the markers
```
