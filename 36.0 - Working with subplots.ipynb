```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# Workingwith subplots
```

```python
sns.set_style('dark')
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRatings)
```

```python
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRatings)
```

```python
f, ax = plt.subplots(1,3, figsize = (12,6))      #plt.subplots(rows,columns, figsize = (size separated by ,))
```

```python
f, axes = plt.subplots(1,2, figsize = (12,6))
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRatings , ax = axes[0])
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRatings, ax = axes[1])
```

```python
# if you want to make more than one dimenssion of charts like 2,2 the behavior of the axes will be like a Matrix
# so in the ax = axes [row, column] like in a matrix

f, axes = plt.subplots(2,2, figsize = (12,6))
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRatings , ax = axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRatings, ax = axes[1,1])
```

```python
f, axes = plt.subplots(1,2, figsize = (12,6),sharex = True, sharey = True) # sharex makes the graphs share the x axe,so when you define for one, is defined for all
k1 = sns.kdeplot(movies.BudgetMillions, movies.AudienceRatings , ax = axes[0])
k2 = sns.kdeplot(movies.BudgetMillions, movies.CriticRatings, ax = axes[1])
k1.set(xlim = (-20,160))  #setting a limit to x dimension in k1 graph
plt.show() # to just show the graph without any further information on the top like warnings etc
```
